Here we define our custom wrapping class. It consists of rendering preprocessing, custom reward system, different hyperparameters initializations, the ability to fix seed/s. It acts as an "interceptor" to the original environment.

```
class CustomSokoban(gym.ObservationWrapper):
  def __init__(self, env, imsize, isColor=True, is_custom_rewards=False, n_actions='all', lamda=5, zeta=0.6, seed=None, env_seeds=[]):
      super().__init__(env)
      self.isColor = isColor
      self.seed = seed
      self.lamda = lamda
      self.zeta = zeta
      self.is_custom_rewards = is_custom_rewards
      self.imsize = (imsize, imsize)
      self.observation_space = gym.spaces.Box(low=0, high=255, shape=(3 if isColor else 1, *self.imsize), dtype=np.uint8)
      self.target = np.array(list(env.box_mapping.keys())[0])
      self.best_distance = self._set_initial_distance()
      self.valid_actions = self._set_valid_actions(n_actions)
      self.env_seeds = env_seeds


  def observation(self, frame):
    frame = frame if self.isColor else cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)  
    frame = cv2.resize(
          frame, self.imsize, interpolation=cv2.INTER_AREA)
    frame = np.expand_dims(frame, 0) if not self.isColor else frame.transpose(2,0,1)
    return torch.tensor(frame, dtype=torch.float32, device=device)

  def get_room_state(self):
    return torch.tensor(super().unwrapped.room_state, device=device, dtype=torch.float32).unsqueeze(0) ** 2

  def step(self, action):
    _, reward, done, info = super().step(action)
    reward = self._generate_custom_reward(reward, done, info)
    reward = torch.tensor(np.array([reward]), device=device, dtype=torch.float32)
    done = torch.tensor([done], device=device)
    next_state = self.observation(super().render('tiny_rgb_array'))

    return next_state, reward, done, info

  def reset(self):
    if len(self.env_seeds) != 0:
      seed = self.env_seeds.pop()
    else:
      seed = self.seed 
    random.seed(seed)
    _ = super().reset()
    random.seed(None)
    state = self.observation(self.env.render('tiny_rgb_array'))
    self.target = np.array(list(self.env.box_mapping.keys())[0])
    self.best_distance = self._set_initial_distance()

    return state

  def _set_initial_distance(self):
    box = np.array(np.where(self.env.unwrapped.room_state == 4)).flatten()
    return np.linalg.norm(self.target - np.array(box), ord=1)

  def _set_valid_actions(self, n_actions):
      try:
        self.valid_actions = action_spaces[n_actions]
      except KeyError:
        raise KeyError(f'n_actions can be one of {action_spaces.keys()}')
      self.action_space.n = len(self.valid_actions)

  def _generate_custom_reward(self, reward, done, info):
    if not self.is_custom_rewards:
      return reward
    if done:
      reward += 5

    elif info['action.moved_box']:
      box = self.new_box_position
      if box is None:
        box = np.array(np.where(self.env.unwrapped.room_state == 4)).flatten()
      distance = np.linalg.norm(self.target - np.array(box), ord=1)
      if distance < self.best_distance:
        reward += self.lamda / (distance)
        self.best_distance = distance

    elif not info['action.moved_player']:
      reward -= self.zeta

    return reward
```

```
def test_agent(agent, env_seeds, model= "DDQN", determin=True):
  steps_history = []
  wins = 0
  for seed in env_seeds: 
    agent.env.seed = seed  
    state = agent.env.reset()
    done = False
    while not done:
      state = agent.env.get_room_state()
      if model == "PPO":
        with torch.no_grad():
          dist, _ = agent.model(state)
          action = dist.sample()
      else:
        action = agent.act(state, determin)
      state, reward, done, _ = agent.env.step(action.item())
    if agent.env.num_env_steps < agent.env.max_steps:
      wins += 1
      steps_history.append(agent.env.num_env_steps)
      
  agent.env.seed = None
  return np.mean(steps_history), wins/len(env_seeds)
```

## Models

### DQN

The default net used by our DQN model. It consists of 2 convolutional layers, followed by 2 Fully connected layers. It supports any amount of action spaces, which is crucial for our action spaces experiments.

```
class Net(nn.Module):
    """
    Convolutional Neural Net with 3 conv layers and two linear layers
    """
    def __init__(self, input_shape, n_actions, conv=None):
        super(Net, self).__init__()
        c, h, w = input_shape

        if conv is not None:
          self.conv = conv
        else:
          self.conv = nn.Sequential(
              nn.Conv2d(c, 4, kernel_size=3, stride=1, padding=1),
              nn.BatchNorm2d(4, momentum=0.01),
              nn.LeakyReLU(),
              nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=0),
              nn.LeakyReLU()
          )
        conv_out_size = self._get_conv_out(input_shape)
        self.fc = nn.Sequential(
            nn.Linear(conv_out_size, 256),
            nn.LeakyReLU(),
            nn.Linear(256, n_actions),
        )
        self.flat = nn.Flatten(1)
    
    def _get_conv_out(self, shape):
        o = self.conv(torch.zeros(1, *shape))
        return int(np.prod(o.size()))

    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.conv(x)
        x = self.flat(x)
        return self.fc(x)
```

Define the replay memory buffer as a queue, using a custom class with the help of a namedtuple object.

```
Transition = namedtuple('Transition',
                        ('state', 'action', 'next_state', 'reward', 'done'))

class ReplayMemory(object):

    def __init__(self, capacity):
        self.memory = deque([], maxlen=capacity)

    def push(self, *args):
        """Save a transition"""
        self.memory.append(Transition(*args))

    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)

    def __len__(self):
        return len(self.memory)
```

Here we define our DQN agent model. All hyper parameters are well documented in the paper. We use smooth L1 loss and *Adam* optimizer.

```
class DQNAgent:
    def __init__(self, env: CustomSokoban, max_memory_size, gamma, lr, epsilon, epsilon_min, epsilon_decay, target_update_freq):
        self.env = env
        self.policy_net = Net(env.observation_space.shape, env.action_space.n).to(device)
        self.target_net = Net(env.observation_space.shape, env.action_space.n).to(device)
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.replay_buffer = ReplayMemory(max_memory_size)
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.optimizer = torch.optim.Adam(self.policy_net.parameters(), lr=lr)
        self.criterion = nn.SmoothL1Loss().to(device)
        self.target_update_freq = target_update_freq
        self.losses = []
        
    def act(self, state, determin=False):
            """Epsilon-greedy action"""
            if random.random() < self.epsilon and not determin:
                return torch.tensor([random.randrange(self.env.action_space.n)], device=device)
            with torch.no_grad():
                return torch.argmax(self.policy_net(state)).view(1)

    def remember(self, state, action, next_state, reward, done):
        """Add the experience to memory"""
        self.replay_buffer.push(state, action, next_state, reward, done)

    def recall(self, batch_size):
        transitions = self.replay_buffer.sample(batch_size)
        batch = Transition(*zip(*transitions))
        state_batch = torch.cat(batch.state).to(device)
        action_batch = torch.cat(batch.action).to(device)
        reward_batch = torch.cat(batch.reward).to(device)
        done_batch = torch.cat(batch.done).to(device)
        next_state_batch = torch.cat(batch.next_state).to(device)
        return state_batch, action_batch, reward_batch, next_state_batch, done_batch

    def learn(self, batch_size):
        """Update online action value (Q) function with a batch of experiences"""
        if len(self.replay_buffer) < self.replay_buffer.memory.maxlen:
            return

        state_batch, action_batch, reward_batch, next_state_batch, done_batch = self.recall(batch_size)
        self.optimizer.zero_grad()
        prime_actions = self.policy_net(next_state_batch).max(1).indices
        next_state_action_value = self.target_net(next_state_batch).gather(1, prime_actions.unsqueeze(1))
        target = (reward_batch + ~done_batch * (self.gamma * next_state_action_value).squeeze()).unsqueeze(1)
        
        state_action_values = self.policy_net(state_batch).gather(1, action_batch.unsqueeze(1))
        loss = self.criterion(state_action_values, target)
        self.losses.append(loss.item())
        loss.backward()
        self.optimizer.step()
        if self.epsilon > self.epsilon_min:
          self.epsilon *= self.epsilon_decay
        
        elif self.env.num_env_steps == self.env.max_steps:
          self.epsilon *= 2

    def do_step(self, state):
      action = self.act(state)
      next_state, reward, done, _ = self.env.step(action.item())
      next_state = self.env.get_room_state()
      self.replay_buffer.push(state, action, next_state, reward, done)
      return next_state, reward, done

    def fill_replay_buffer(self, verbose):
      if verbose:
        print("Filling Replay Buffer...")
      while len(self.replay_buffer) < self.replay_buffer.memory.maxlen:
        state = self.env.reset()
        state = self.env.get_room_state()
        while True:
          next_state, reward, done = self.do_step(state)
          state = next_state
          if done:
            break
      if verbose:
        print("Replay Buffer is Full, start learning")

    def run(self, episodes, batch_size=32, verbose=True):
      print_iter = 50 if verbose is True else 1
      self.fill_replay_buffer(verbose)   
      total_steps = 0 
      results = defaultdict(list)  
      for episode in range(1, episodes+1):
        rewards = 0
        state = self.env.reset()
        state = self.env.get_room_state()
        while True:
          next_state, reward, done = self.do_step(state)
          total_steps += 1 
          rewards += reward
          state = next_state
          self.learn(batch_size)
          
          if total_steps % self.target_update_freq == 0:
             self.target_net.load_state_dict(self.policy_net.state_dict()) 
          if done:
            results["rewards"].append(rewards.item())
            results["wins"].append(int(self.env.num_env_steps < self.env.max_steps))
            results["steps"].append(self.env.num_env_steps)
            break

        if verbose and episode % print_iter == 0:
          print(f'Episode: {episode}, finished with {self.env.num_env_steps} steps and total rewards of {rewards.item()}')
      return results
```

### PPO

The cell below consists of the definitions of both the actor and critic's neural networks, with the ability to inject different convolutional structures.

```
class ActorCritic(nn.Module):
    def __init__(self, input_shape, n_actions, critic_conv=None, actor_conv=None, hidden_size=256):
        super(ActorCritic, self).__init__()
        c, h, w = input_shape
        if critic_conv != None:
          self.critic_conv = critic_conv
        else:
          self.critic_conv = nn.Sequential(
              nn.Conv2d(c, 4, kernel_size=3, stride=1, padding=1),
              nn.BatchNorm2d(4, momentum=0.01),
              nn.LeakyReLU(),
              nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=0),
              nn.LeakyReLU()
            )

        conv_out_size= self._get_conv_out(input_shape, self.critic_conv)
        self.critic_fc = nn.Sequential(
            nn.Linear(conv_out_size, hidden_size),
            nn.LeakyReLU(),
            nn.Linear(hidden_size, 1)
        )
        if actor_conv != None:
          self.actor_conv = actor_conv
        else:
          self.actor_conv = nn.Sequential(
              nn.Conv2d(c, 4, kernel_size=3, stride=1, padding=1),
              nn.BatchNorm2d(4, momentum=0.01),
              nn.LeakyReLU(),
              nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=0),
              nn.LeakyReLU()
            )
        conv_out_size= self._get_conv_out(input_shape, self.actor_conv)
        self.actor_fc = nn.Sequential(
            nn.Linear(conv_out_size, hidden_size),
            nn.LeakyReLU(),
            nn.Linear(hidden_size, n_actions),
            nn.Softmax()
        )
        self.flatten = nn.Flatten(1)

    def _get_conv_out(self, shape, conv):
      o = conv(torch.zeros(1, *shape))
      return int(np.prod(o.size())) 

    def forward(self, x):
      x = x.unsqueeze(1)
      out_critic = self.critic_conv(x)
      out_critic = self.flatten(out_critic)
      value = self.critic_fc(out_critic)
      out_actor = self.actor_conv(x)
      out_actor = self.flatten(out_actor)

      mu = self.actor_fc(out_actor)
      dist  = Categorical(mu)
      return dist, value
```

Below is our PPO model implementation. The hyperparameters are documented in the paper. We use *Adam* optimizer similarly to DDQN.

```
class PPO():
    def __init__(self, env, model, lr=1.5e-4) -> None:
        self.env = env
        self.test_rewards = []
        self.model = model.to(device)
        self.optimizer =  torch.optim.Adam(self.model.parameters(), lr=lr)
        self.total_episodes = 0

    def run(self, episodes, num_steps=20, mini_batch_size=5, ppo_epochs=4, tau=0.95, verbose=True):
      results = defaultdict(list)
      for episode in range(1, episodes+1):
          log_probs = []
          values    = []
          states    = []
          actions   = []
          rewards   = []
          masks     = []
          entropy = 0
          episode_reward = 0
          state = self.env.reset()
          while True:
              for _ in range(num_steps):
                  state = self.env.get_room_state()
                  dist, value = self.model(state)

                  action = dist.sample()
                  next_state, reward, done, _ = self.env.step(action.item())

                  log_prob = dist.log_prob(action)
                  entropy += dist.entropy().mean()
                    
                  log_probs.append(torch.tensor([log_prob]).unsqueeze(1))
                  values.append(value)
                  rewards.append(reward.unsqueeze(1))
                  masks.append(~done.unsqueeze(1))
                    
                  states.append(state)
                  actions.append(action)
                  #added next line
                  next_state = self.env.get_room_state()
                  state = next_state
                  episode_reward += reward.item()
                  if done:
                      break

              _, next_value = self.model(next_state)
              returns = self._compute_gae(next_value, rewards, masks, values, tau)

              returns_tensor   = torch.cat(returns).detach().to(device)
              log_probs_tensor = torch.cat(log_probs).detach().to(device)
              values_tensor    = torch.cat(values).detach().to(device)
              states_tensor    = torch.cat(states).to(device)
              actions_tensor  = torch.tensor(actions).unsqueeze(1).to(device)
              advantage_tensor= returns_tensor - values_tensor
                
              self.ppo_update(ppo_epochs, mini_batch_size, states_tensor, actions_tensor, log_probs_tensor, returns_tensor, advantage_tensor)
            
              if done:
                results["rewards"].append(episode_reward)
                results["wins"].append(int(self.env.num_env_steps < self.env.max_steps))
                results["steps"].append(self.env.num_env_steps)
                break

          if episode % 10 == 0 and verbose:
            print(f'Episode: {episode}, finished with {self.env.num_env_steps} steps and total rewards of {episode_reward}')     
      return results

    def ppo_update(self, ppo_epochs, mini_batch_size, states, actions, log_probs, returns, advantages, clip_param=0.2):
        for _ in range(ppo_epochs):
            for state, action, old_log_probs, return_, advantage in self._ppo_iter(mini_batch_size, states, actions, log_probs, returns, advantages):
                dist, value = self.model(state)
                entropy = dist.entropy().mean()
                new_log_probs = dist.log_prob(action.T).T

                ratio = (new_log_probs - old_log_probs).exp()
                surr1 = ratio * advantage
                surr2 = torch.clamp(ratio, 1.0 - clip_param, 1.0 + clip_param) * advantage

                actor_loss  = - torch.min(surr1, surr2).mean()
                critic_loss = 0.5*torch.mean((return_ - value)**2)
                loss = critic_loss + actor_loss - 0.001 * entropy

                self.optimizer.zero_grad()
                loss.backward()
                self.optimizer.step()

    def _ppo_iter(self, mini_batch_size, states, actions, log_probs, returns, advantage):
        batch_size = states.size(0)
        for _ in range(batch_size // mini_batch_size):
            rand_ids = np.random.randint(0, batch_size, mini_batch_size)
            yield states[rand_ids, :], actions[rand_ids, :], log_probs[rand_ids, :], returns[rand_ids, :], advantage[rand_ids, :]

    def _compute_gae(self, next_value, rewards, masks, values, tau, gamma=0.99):
      values = values + [next_value]
      gae = 0
      returns = []
      for step in reversed(range(len(rewards))):
          delta = rewards[step] + gamma * values[step + 1] * masks[step] - values[step]
          gae = delta + gamma * tau * masks[step] * gae
          returns.insert(0, gae + values[step])
      return returns
        
```

#### Experiment 3 - Different Net Architectures 

Below are the different net architectures we use during the experiment, stored in a dictionary for easy documentation.

```
net1 = nn.Sequential(
  nn.Conv2d(env_ex2.observation_space.shape[0], 8, kernel_size=3, stride=1, padding=1),
  nn.BatchNorm2d(8, momentum=0.01),
  nn.LeakyReLU(),
  nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=0),
  nn.LeakyReLU()
)

net2 = nn.Sequential(
  nn.Conv2d(env_ex2.observation_space.shape[0], 4, kernel_size=3, stride=1, padding=1),
  nn.BatchNorm2d(4, momentum=0.01),
  nn.LeakyReLU(),
  nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=0),
  nn.LeakyReLU()
)
net3 = nn.Sequential(
    nn.Conv2d(env_ex2.observation_space.shape[0], 4, kernel_size=3, stride=1, padding=1),
    nn.BatchNorm2d(4, momentum=0.01),
    nn.LeakyReLU(),
    nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=1),
    nn.LeakyReLU(),
    nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=0),
    nn.LeakyReLU(),
)
net4 = nn.Sequential(
    nn.Conv2d(env_ex2.observation_space.shape[0], 4, kernel_size=3, stride=1, padding=1),
    nn.BatchNorm2d(4, momentum=0.01),
    nn.LeakyReLU(),
    nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=1),
    nn.LeakyReLU(),
    nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=0),
    nn.LeakyReLU(),
    nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0),
    nn.LeakyReLU(),
)
net5 = nn.Sequential(
    nn.Conv2d(env_ex2.observation_space.shape[0], 512, kernel_size=7, stride=1, padding=0),
    nn.LeakyReLU()
)
net6 = nn.Sequential(
    nn.Conv2d(env_ex2.observation_space.shape[0], 4, kernel_size=2, stride=1, padding=0),
    nn.BatchNorm2d(4, momentum=0.01),
    nn.LeakyReLU(),
    nn.Conv2d(4, 8, kernel_size=2, stride=1, padding=0),
    nn.LeakyReLU(),
    nn.Conv2d(8, 16, kernel_size=2, stride=1, padding=0),
    nn.LeakyReLU(),
)
net7 = nn.Sequential(
    nn.Conv2d(env_ex2.observation_space.shape[0], 4, kernel_size=3, stride=1, padding=1),
    nn.BatchNorm2d(4, momentum=0.01),
    nn.LeakyReLU(),
    nn.Conv2d(4, 8, kernel_size=3, stride=1, padding=1),
    nn.BatchNorm2d(8, momentum=0.01),
    nn.LeakyReLU(),
    nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=0),
    nn.BatchNorm2d(16, momentum=0.01),
    nn.LeakyReLU(),
)
nets = {"Deep": net1, "2-layers": net2, "3-layers": net3, "4-layers": net4, "Fc": net5, "kernel-2": net6, "BN": net7}
```

Note that we initialize the optimizer again outside the DQNAgent class to include the injected net weights in the backpropogation process as well.

```
net_results = {}
nets_evals = defaultdict(dict)
for net_name, net in nets.items():
  env_ex2 = CustomSokoban(deepcopy(sok_dqn), 7, isColor=False, is_custom_rewards=True, env_seeds=seeds.copy())
  agent_nets = DQNAgent(env_ex2, max_memory_size=10000, gamma=0.99, lr=0.00025, epsilon=1, epsilon_min=0.05, epsilon_decay=0.99985, target_update_freq=50)
  agent_nets.replay_buffer = deepcopy(replay_buffer)
  agent_nets.policy_net = Net(env_ex2.observation_space.shape, env_ex2.action_space.n, deepcopy(net)).to(device)
  agent_nets.target_net = Net(env_ex2.observation_space.shape, env_ex2.action_space.n, deepcopy(net)).to(device)
  agent_nets.target_net.load_state_dict(agent_nets.policy_net.state_dict())
  agent_nets.optimizer = torch.optim.Adam(agent_nets.policy_net.parameters(), lr=0.00025)
  print(f"Running agent with {net_name} net...")
  results = agent_nets.run(400, verbose=False)
  mean_steps, win_ratio = test_agent(agent_nets, test_seeds)
  nets_evals["Average_Steps"][str(net_name)] = mean_steps
  nets_evals["Success_Rate"][str(net_name)] = win_ratio
  net_results[net_name] = results
```

```
nets_values = get_extreme_values(net_results, nets_evals)
draw_rewards_and_wins_graph(nets_values)
```
