#### 1.F
Data with τ(x,y) < ρ(x,y) – 1.2 

We think that this case is impossible since the two correlation methods (Spearman and Kendall) are based on the rank's order and not the exact values of the variables. both behave almost the same base on the monotonic relationship. <br/>
If we take for example Spearman = 1 and Kendall = -2, in this case according to Spearman the data is monotonically increasing which mean that all data pairs should be concordant and Kendall will also equal 1. <br/>
If we take Kendall = -1 so Spearman = 0.2, in this case all the data pairs will be discordant and are monotonically decreasing so Spearman should also be -1.
Therefore, we cannot provide any example in which the Spearman correlation result will be 1.2 higher than the Kendall correlation, which means we will get a completely different correlation. <br/>
Our intuition is that the maximum difference between the two correlation can be at most 0.5 as n goes to infinity (as we tried brute-force with data samples while raising n and only be able to get near 0.5).

```python
# Example code for bivariate pdfs
from scipy.stats import multivariate_normal as mn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2d_Gaussian_pdf(means, cov, title, x_label, y_label):
    n = 100
    x1 = np.linspace(means[0] - 2 * np.sqrt(cov[0][0]), means[0] + 2 * np.sqrt(cov[0][0]), n)
    x2 = np.linspace(means[1] - 2 * np.sqrt(cov[1][1]), means[1] + 2 * np.sqrt(cov[1][1]), n)
    x1_v, x2_v = np.meshgrid(x1, x2)
    Xgrid = np.vstack([x1_v.ravel(), x2_v.ravel()]).T
    Y = mn.pdf(Xgrid, means, cov)
    fig, ax = plt.subplots()
    ax.pcolorfast(x1, x2, Y.reshape(x1_v.shape), alpha=0.5, cmap='Blues')
    ax.contour(x1_v, x2_v, Y.reshape(x1_v.shape), alpha=0.3, colors='b')
    ax.set(title=title, xlabel=x_label, ylabel=y_label)
    ax.grid(alpha=0.2)
    plt.show()
```

```python
numeric_attributes = ['age', 'trestbps', 'chol', 'thalach','oldpeak']
categorical_attributes = ['sex', 'cp', 'restecg', 'exang', 'slope', 'ca', 'thal', 'fbs']

for c in numeric_attributes:
    df[c].fillna(df[c].median(), inplace=True)

for c in categorical_attributes:
    df[c].fillna(df[c].value_counts().index[0], inplace=True)
print(df.info())
```

```python
df['target'].replace([0,1],['healthy','disease'],inplace=True )
df['sex'].replace([0,1],['female','male'],inplace=True )
```

```python
alpha = 0.05
z_critical = norm.ppf(q = 1-alpha/2) 
f = lambda x: (round(x.mean() - z_critical* x.std(ddof=1), 3), round(x.mean() + z_critical*x.std(ddof=1), 3))
df1 = df.drop(['sex', 'cp','fbs','restecg','exang','slope','ca','thal'], axis=1).groupby('target').agg(f)
intervals = df1.T
intervals
```

```python
df_healthy = df.loc[df['target'] == 'healthy']
df_disease = df.loc[df['target'] == 'disease']
n1, n2 = len(df_disease), len(df_healthy)
alpha = 0.05

for c in numeric_attributes:
    m_disease = df_disease[c].mean()
    s_disease = df_disease[c].std(ddof=1)
    m_healthy = df_healthy[c].mean() 
    s_healthy = df_healthy[c].std(ddof=1)
    SE = np.sqrt(s_disease**2 / n1 + s_healthy**2 / n2)
    low, high = (m_healthy - m_disease) - SE*norm.ppf(1-alpha/2), (m_healthy-m_disease) + SE*norm.ppf(1-alpha/2)

    print(f"\033[1m{c}\033[0m:")
        
    print(f"Healthy mean: {m_healthy:.4f} Disease mean: {m_disease:.4f}")
    print(f'({low:.3f}, {high:.3f})')
    
    if low <= 0 <= high:
        print(f"Not enough confidence to reject the null hypothesis (that the healthy pop mean = disease pop mean)")
    else:
        sign = ['smaller', 'larger']
        print(f"In confidence of 95% the healthy pop mean is *{sign[m_healthy > m_disease]}* than the disease pop mean")
    print()
```

Observing the plot, the histogram appears to be in the shape of a Normal distribution.
by using MLE we derive the disribution parameters:

$$L(\Theta)=ln(p(D\mid\Theta)=ln(\varphi(x_{1},x_{2},...x_{n}\mid\Theta)=ln(\prod_{i}^{n}\varphi(x_{i}\mid\Theta)$$

$$=\sum_{i}^{n}ln(\varphi(x_{i}\mid\Theta))=\sum_{i}^{n}ln(\frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-(x_{i}-\mu)^{2}}{2\sigma^{2}}})=$$

$$\sum_{i}^{n}-\frac{1}{2}ln(2\pi\sigma^{2})-\frac{(x_{i}-\mu)^{2}}{2\sigma^{2}}=$$

$$\frac{dL(\Theta)}{d\mu}=-\frac{1}{2\sigma^{2}}\sum_{i}^{n}-2(x_{i}-\mu)=0$$

$$\hat{\mu}=\frac{1}{n}\sum_{i}^{n}x_{i}$$

Similarly we do the same for σ:

$$\sum_{i}^{n}-\frac{1}{2}ln(2\pi\sigma^{2})-\frac{-(x_{i}-\mu)^{2}}{2\sigma^{2}}=$$

$$\frac{n}{2}ln(2\pi\sigma^{2})-\frac{1}{2\sigma^{2}}\sum_{i}^{n}(x_{i}-\mu)^{2}$$

$$\frac{dL(\Theta)}{d\sigma}=-\frac{n}{\sigma}+\frac{1}{\sigma^{3}}\sum_{i}^{n}(x_{i}-\mu)^{2}=0$$

$$\sum_{i}^{n}(x_{i}-\mu)^{2}=n\sigma^{2}$$

$$\hat{\sigma=}\sqrt{\frac{1}{n}\sum_{i}^{n}(x_{i}-\mu)^{2}}$$

So we can used the ampirical μ and σ to fit the Normal curve.

```python
best_distributions = dict()
    
for feature in numeric_attributes:
    feature_data = df[feature]
    max_mle_score = None
    best_dist = None
    for dist in continuous_distributions:
        score = np.sum(np.log(dist.pdf(feature_data, *dist.fit(feature_data))))
        if best_dist is None or score >= max_mle_score:
            max_mle_score = score
            best_dist = dist
        best_distributions[feature] = (best_dist.name, best_dist) 
```

```python
for f in numeric_attributes:
    plt.figure(figsize=(14,4))
    dist_name, dist = best_distributions[f]
    plt.title(f"{f} histogram - fitted by {dist_name} distribution")
    sn.distplot(df[f], kde=False, norm_hist=True, bins=30, fit=dist)
    plt.show()
```

As such, the null hypothesis may state that the male probability is smaller or equal than the female probability, and the alternative hypothesis will state that the male probability is greater than the female probability.

Based on the results below, the lower limit of the proportion difference interval is above zero, so we can state that male proportion among subject with heart disease is statistically significantly bigger that female proportion with significance confidence of 95%, therefore we can reject the null hypothesis. 

```python
df_disease = df[df['target'] == 'disease']
n = df_disease.shape[0]
p_male = df_disease[df_disease['sex'] == 'male']['sex'].count() / n
p_female = df_disease[df_disease['sex'] == 'female']['sex'].count() / n

print(f'The male empirical proportion: {p_male:.4f}')
print(f'The female empirical proportion: {p_female:.4f}')

alpha = 0.05
gamma = norm.ppf(q=1-alpha/2)

sigma_male = np.sqrt(p_male*(1-p_male)) / np.sqrt(n)
left_male = p_male - gamma * sigma_male
right_male = p_male + gamma * sigma_male
male_interval = (left_male, right_male)
print(f'Confidence Interval for the male proportion: {male_interval}')

sigma_female = np.sqrt(p_female*(1-p_female)) / np.sqrt(n)
left_female = p_female - gamma * sigma_female
right_female = p_female + gamma * sigma_female
female_interval = (left_female, right_female)
print(f'Confidence Interval for the female proportion: {female_interval}')

empirical_delta = p_male - p_female
error_margin_delta = gamma * np.sqrt(((p_male*(1-p_male))/n) +
                                     ((p_female*(1-p_female))/n))

left_delta = empirical_delta - error_margin_delta
right_delta = empirical_delta + error_margin_delta
delta_interval = (left_delta, right_delta)
print(f'Confidence Interval for the difference between gender proportion: {delta_interval}')
```

Calculate $P(X_1 < X_2)$.

$$1-\int_{0}^{\infty}f_{1}(x)\left(\int_{0}^{x}f_{2}(y)dy)\right)dx=$$

```python
def covariance(index, pi_0, states, T):
    X_mean, Y_mean, XY_mean = 0, 0, 0
    
    T_i = matrix_power(T, index)
    pi_i = pi_0 @ T_i

    #E(X0)
    X_mean  = P0 @ states.T

    #E(Xi)
    Y_mean += pi_i @ states.T

    #E(X_0 * Xi)
    for x in states:
        for y in states:
            XY_mean += x * y * T_i[x - 1, y - 1] * pi_0[x-1]

    return XY_mean - X_mean * Y_mean

states = np.arange(1,7)
P0 = np.ones(6)/6
covariances = np.zeros(500)
for i in range(0,500):
  covariances[i]=covariance(i+1, P0, states, T )
print(covariances)
```
